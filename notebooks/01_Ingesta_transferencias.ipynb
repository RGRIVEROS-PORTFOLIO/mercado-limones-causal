## Hallazgos — Lo que la serie continua revela

### 1. Estacionalidad de diciembre — patrón estructural confirmado

Cada diciembre es el mínimo de antigüedad de su año:
- 2023-12: 14.46 → 2024-01: 15.13 (rebote)
- 2024-12: 13.34 → 2025-01: 14.39 (rebote)
- 2025-12: 13.64 → 2026-01: 14.47 (rebote)

Consistente con Jensen & Meckling (1976): presión de objetivos comerciales
de fin de año fuerza liquidación de mejores unidades. No es evento aislado
— es regularidad estructural del mercado. Requiere variable dummy estacional
en el modelo ADL.

### 2. Epicentro del shock Milei — enero 2024, no febrero

2024-01: 15.13 años es el punto más alto de toda la serie.
El diseño muestral del proyecto anterior había elegido 2024-02 como proxy
del shock, pero la serie continua muestra que el pico fue un mes antes.
**Implicancia para el modelo causal:** el lag real del shock puede ser
distinto al estimado con 7 puntos.

### 3. Protocolo 21 — tendencia ascendente secular

Participación promedio por año:
- 2023: ~40.2%
- 2024: ~39.3% (baja transitoria durante shock)
- 2025: ~41.3%
- 2026 (ene-may): ~42.1%

La desregulación de importaciones post-2024 tiene huella clara y sostenida.
Esto refuerza la necesidad de variable de control macroeconómico en el modelo
para separar efecto desregulación del efecto selección adversa.
